```python
for i in range(len(data.columns)):
    print(data.iloc[:, i].describe(),"\n")
```

```python
##### SUPPORT FUNCTIONS
def unique_vals(idx, counts=True):
    print(f"Unique values for '{data.columns[idx]}':\n")
    for value in data.iloc[:, idx].unique():
        print(value)
    print()
    
    if counts: print(data.iloc[:, idx].value_counts())
```
